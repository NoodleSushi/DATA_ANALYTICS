```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
computer_hardware = fetch_ucirepo(id=29) 
  
# data (as pandas dataframes) 
X = computer_hardware.data.features 
y = computer_hardware.data.targets 
  
# metadata 
# print(computer_hardware.metadata) 
  
# variable information 
print(computer_hardware.variables) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Plot the distribution of machine cycle time
plt.figure(figsize=(10, 6))
sns.histplot(X['MYCT'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Machine Cycle Time')
plt.xlabel('Machine Cycle Time')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of vendors
plt.figure(figsize=(10, 6))
sns.countplot(x='VendorName', data=X, palette='viridis')
plt.title('Distribution of Vendors')
plt.xlabel('Vendor')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()





# Plot the distribution of vendors
plt.figure(figsize=(12, 8))
occupation_counts = X['VendorName'].value_counts()
plt.pie(occupation_counts, labels=occupation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Vendors')
plt.show()


# Plot the relationship between published relative performance and estimated relative performance
plt.figure(figsize=(12, 8))
sns.violinplot(x='VendorName', y='ERP', data=X, palette='pastel')
plt.title('Vendor vs Estimated Relative Performance (ERP)')
plt.xlabel('Vendor')
plt.ylabel('ERP')
plt.xticks(rotation=4)
plt.show()
```
